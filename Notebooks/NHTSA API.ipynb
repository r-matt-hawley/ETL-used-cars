```python
import requests, json
from pprint import pprint
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetVehicleVariableList?format=json';
r = requests.get(url)
print(r.json())
```

```python
pprint(r.json())
```

```python
variable_list = []
index_list = []
for result in r.json()["Results"]:
    variable_list.append(result["Name"])
    index_list.append(result["ID"])
```

```python
import pandas as pd
```

```python
variable_df = pd.DataFrame({"Variable":variable_list}, index=index_list)
variable_df.head()
```

```python
sorted_df = variable_df.sort_values("Variable")
for name in sorted_df["Variable"]:
    print(name)
```
