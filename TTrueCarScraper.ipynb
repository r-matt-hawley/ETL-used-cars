```python
url = "https://www.truecar.com/used-cars-for-sale/listings/year-2017-max/location-houston-tx/?searchRadius=100&sort[]=price_desc"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

nextt = soup.find_all('svg', attrs={"data-qa": "IconAngleRight"})

autos = nextt.find_all('svg')

category_list = []
url_list = []
book_url_list = []
for category in autos:
    title = category.text.strip()
    category_list.append(title)
    book_url = category.find('a')['href']
    url_list.append(book_url)

TrueCar_url_list = ["https://www.truecar.com/used-cars-for-sale/listings/year-2017-max/location-houston-tx/?searchRadius=100&sort[]=price_desc" + url for url in url_list]

titles_and_urls = zip(category_list, TrueCar_url_list)

try:
    for title_url in titles_and_urls:
        browser.click_link_by_partial_text('next')
except ElementDoesNotExist:
    print("Scraping Complete")
```
